```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from matplotlib import rcParams
rcParams['font.family'] = 'sans-serif'
rcParams['font.sans-serif'] = ['Helvetica']
rcParams['pdf.fonttype'] = 42
```

```python
df = pd.read_excel('data/permis-acquisition.xlsx', index_col=0)
```

```python
_df = df.copy()
```

```python
df['delta'] = df[2018] / df[2014]
df['delta']
```

```python
cantons_order = df.sort_values('delta', ascending=False).index.tolist()
```

```python
years = _df.columns

plt.style.use('seaborn-darkgrid')
plt.figure(figsize=(14, 14))
palette = plt.get_cmap('Set1')
num = 1

def getColor(num):
#    return palette(num)
    return '#c00d0e'

for name in cantons_order:
    row = _df.loc[name]
    
    column = name
 
    plt.subplot(5, 2, num)
 
    plt.plot(years, row, marker='', color=getColor(num), linewidth=2.4, alpha=0.9, label=column)
 
    # Add title
    plt.title(column, loc='left', fontsize=16, fontweight=800, color = getColor(num))
    plt.locator_params(axis='x', nbins=5)

    num += 1

plt.tight_layout()
plt.savefig('small-multiples.pdf')
```
